```python
import numpy as np; import matplotlib.pyplot as plt; import pandas as pd; import cmath; import math; import sys; plt.rcParams['figure.figsize']= [20,8]

my_file = open("NewHunt_21D070021.dat", "r")
other_file = open("NewHunt_21D070053.dat", "r")
  
data = my_file.read()
other_data = other_file.read()
  
data_into_list = data.split("\n")
other_data_into_list = other_data.split("\n")

my_file.close()
other_file.close()

complex_num_list = []
complex_num_list_1 = []

for i in range(0,len(data_into_list)-1):
  complex_num_list.append(complex(data_into_list[i]))
  complex_num_list_1.append(complex(other_data_into_list[i]))

#def intersection(real_values, real_values_1):
    #real_vals = [value for value in real_values if value in real_values_1]
    #return real_vals

real_values = []; real_values_1 = [] 
imag_values = []; imag_values_1 = []
indices = []; indices_1 = []
newList = []; newList_1 = []

for i in range(0,len(complex_num_list)):
  real_values.append(complex_num_list[i].real); real_values_1.append(complex_num_list_1[i].real)
  imag_values.append(complex_num_list[i].imag); imag_values_1.append(complex_num_list_1[i].imag)
  indices.append(i)
  indices_1.append(len(complex_num_list)-1-i)

for element in real_values:
    if element in real_values_1:
        newList.append(element)

plt.scatter(indices, real_values, s=5, c='blue')  
plt.scatter(indices, real_values_1, s=5, c='red')
plt.xlabel('Indices of the data points')
plt.ylabel('Real part of the data points')

print('The number of elements common to both the lists are: ' + str(len(newList)))
indices_common = [i for i, real_values_ in enumerate(real_values) if real_values_ in real_values_1]
print('The indices of elements (belonging to my data set) that are common to the uploaded data set are: ' + str(indices_common))

plt.show()
```
